```python
def count_golas(s): 
    s = s.split('–')
    return int(s[0]) + int(s[1])
```

```python
df['Counter_goals'] = df['Score'].apply(count_golas)
```

```python
df
```

```python
def winners_goals(s): 
    s = s.split('–')
    return int(s[0])
```

```python
df["Winners_goals"] = df['Score'].apply(winners_goals)
```

```python
def losers_goals(s): 
    s = s.split('–')
    return int(s[1])
```

```python
df["losers_goals"] = df['Score'].apply(losers_goals)
```

```python
df
```

```python
all_winners = df['Winners'].unique()
```

```python
all_runners = df['Runners-up'].unique()
```

```python
all_teams = np.append(all_winners,all_runners)
```

```python
teams = set(all_teams)
```

```python
teams = np.array(list(teams))
teams
```

```python
len(teams)
```

```python
teams_golas={}
def goals_scored():
    for i in teams :
        h = df[df['Winners']==i]['Winners_goals'].sum()
        a = df[df['Runners-up']==i]['losers_goals'].sum()
        teams_golas.update({i:h+a})
goals_scored()
```

```python
teams_golas
```

```python
a = sorted(teams_golas.items(), key=lambda x: x[1],reverse=True)
a
```

```python
get_goals_teams = {}
def get_goals(): 
    for i in teams: 
        h = df[df['Winners']==i]['losers_goals'].sum()
        a = df[df['Runners-up']==i]['Winners_goals'].sum()
        get_goals_teams.update({i:h+a})
get_goals()
```

```python
def win_percentage():
    for i in teams:
        w = df[df['Winners']==i].shape[0]
        l = df[df['Runners-up']==i].shape[0]
        print(f"{i} reach the final: {w+l},  won in: {w}  and lost in: {l}  percntage to win  the ucl is {w/(w+l)*100}")
        print()
win_percentage()
```
